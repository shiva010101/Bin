```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/winequality-white.csv",sep=";")
col = data.columns
col.tolist()
x_cols = [con for con in col]
x_cols.remove("quality")

for con in x_cols:
    y = np.array([data["quality"].to_list()])
    x = np.array([data[con].to_list()])
    plt.xlabel(con)
    plt.ylabel("quality")
    # plt.plot(x[0],y[0])
    plt.scatter(x[0],y[0])
    print("\n")
    plt.show()
```
